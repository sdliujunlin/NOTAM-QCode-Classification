```python
# 1. 导入必备库
import pandas as pd
from sklearn.model_selection import train_test_split

# 2. 读取Excel文件（修改为你的桌面文献路径）
df = pd.read_excel(
    "C:\\Users\\sdliu\\Desktop\\文献\\NOTAM_test_processed.xlsx",  # Windows路径写法
    usecols=["e_text", "q_code"]  # 若列名不同，仍需替换为你的实际列名
)

# 3. 第一次划分：训练集（80%）+ 临时集（20%）
train_df, temp_df = train_test_split(
    df, 
    test_size=0.2, 
    random_state=42, 
    stratify=df["q_code"]  # 按Q码比例划分，确保类别均匀
)

# 4. 第二次划分：验证集（10%）+ 测试集（10%）
val_df, test_df = train_test_split(
    temp_df, 
    test_size=0.5, 
    random_state=42, 
    stratify=temp_df["q_code"]
)

# 5. 保存3个CSV文件（修改为你的桌面文献路径，方便后续找到文件）
train_df.to_csv("C:\\Users\\sdliu\\Desktop\\文献\\train_data.csv", index=False, encoding="utf-8")
val_df.to_csv("C:\\Users\\sdliu\\Desktop\\文献\\val_data.csv", index=False, encoding="utf-8")
test_df.to_csv("C:\\Users\\sdliu\\Desktop\\文献\\test_data.csv", index=False, encoding="utf-8")

# 6. 输出划分结果核对
print("数据集划分完成！")
print(f"总数据量：{len(df)} 行")
print(f"训练集：{len(train_df)} 行（{len(train_df)/len(df)*100:.1f}%）")
print(f"验证集：{len(val_df)} 行（{len(val_df)/len(df)*100:.1f}%）")
print(f"测试集：{len(test_df)} 行（{len(test_df)/len(df)*100:.1f}%）")
print("\n各数据集Q码类别分布（确保均匀）：")
print("训练集Q码分布：")
print(train_df["q_code"].value_counts().sort_index())
print("\n验证集Q码分布：")
print(val_df["q_code"].value_counts().sort_index())
print("\n测试集Q码分布：")
print(test_df["q_code"].value_counts().sort_index())
```
